# numerical column-------->missing values------take median

```python
#Impute numerical columns with the median value
dataset = pd.DataFrame(dataset)
dataset_cleaned = dataset.copy()
numerical_columns = ['age','bp','sg','al','su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']
for col in numerical_columns:
    dataset_cleaned[col] = dataset_cleaned[col].fillna(dataset_cleaned[col].median())
    print(dataset_cleaned)

```

```python
print(dataset_cleaned[numerical_columns].isnull().sum())
```

```python
dataset = pd.DataFrame(dataset)
dataset['pcv'] = dataset['pcv'].fillna(dataset['pcv'].median())
dataset['wc'] = dataset['wc'].fillna(dataset['wc'].median())
dataset['rc'] = dataset['rc'].fillna(dataset['rc'].median())
missing_values = dataset.isnull().sum()
print("Missing values in each column after imputation:")
print(missing_values)

```

```python
#Drop columns with missing values
dataset_cleaned = dataset.drop(columns=['pcv', 'wc', 'rc'])

#Check the result
print(dataset_cleaned.columns)
```

# categorical column--------->take--->Mode

```python
categorical_columns = ['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane']  # Add other categorical columns here
#Fill missing values in all categorical columns with the mode
for col in categorical_columns:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])
#Check the result
print("\nUpdated DataFrame with missing values filled:")
print(dataset)
```
